### House Prices Prediction

#### Reading the training set

```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load in 

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

#read the train data
train = pd.read_csv('../input/train.csv')

train.shape
```

#### Description of the training set

```python
#Find the info on categorical and continuous variables
train.describe(include=['O'])
```

```python
train.describe()
```

#### Dropping all the columns which have null values more than 30% of total size of data set

```python
#Drop the columns having null values more than 30% of size of dataset
train1 = train.copy()
for column in train1.columns :
    if train1[column].isnull().sum() > 500:
        train1 = train1.drop([column],axis=1)
train1.shape
```

#### Filling the Categorical columns missing values with the mode of the variable and numerical with the mean

```python
#Impute the columns having lesser number of missing values
for column in train1.columns :
    if train1[column].isnull().sum() < 500 and train1[column].isnull().sum() != 0:
        #print(column)
        if train1[column].dtype == "object":
            #print(train1[column].dtype)
            temp = train1[column].mode()
            train1[column] = train1[column].fillna(value = temp[0])
        else:
            train1[column] = train1[column].fillna(value = train1[column].mean())
```

#### Target Variable

```python
#Obtaining the target variable
train_Y = train1.SalePrice
```

```python
#The train set
train1 = train1.drop(['SalePrice'],axis = 1)
```

#### Original encoding the target variables

```python
for variable in train1.columns: # Loop through all columns in the dataframe
    if train1[variable].dtype == 'object': # Only apply for columns with categorical strings
        train1[variable] = pd.Categorical(train1[variable]).codes # Replace strings with an integer
train1.head()
```

#### Splitting the data into train and test for model analysis

```python
#Splitting the train and test data set using train_test_split
from sklearn.model_selection import train_test_split

train_X, test_X, train_y, test_y = train_test_split(train1,train_Y, test_size = 0.20, random_state = 47)
```

#### Decision Tree Regressor Model

```python
#using decision tree for prediction
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error
def getMae(max_leaf_nodes, train_X,test_X, train_Y, test_Y):
    model = DecisionTreeRegressor()
    
    model.fit(train_X, train_Y)
    predicted_values = model.predict(test_X)
    mae = mean_absolute_error(test_Y,predicted_values )
    return mae
```

```python
for leaf in [5,50,500,5000]:
    mae = getMae(leaf,train_X,test_X, train_y, test_y)
    print("leaf : %d \t\t Mae : %d" %(leaf,mae))
```

#### Random Forest Model

```python
#using random forest to do the prediction
from sklearn.ensemble import RandomForestRegressor
def getMaeforRandomForest(train_X,test_X, train_Y, test_Y):
    model = RandomForestRegressor()
    
    model.fit(train_X, train_Y)
    predicted_values = model.predict(test_X)
    mae = mean_absolute_error(test_Y,predicted_values )
    return mae
```

```python
mae = getMaeforRandomForest(train_X,test_X, train_y, test_y)
print("random Forest Mae : %d" %(mae))
```

#### XGBoost Model

```python
#Gradient Boosting for prediction
from xgboost import XGBRegressor
def getMaeforXGBoost(n_estimators,train_X,test_X, train_Y, test_Y):
    model = XGBRegressor(n_estimators = n_estimators)
    
    model.fit(train_X, train_Y, early_stopping_rounds=5, eval_set=[(test_X, test_Y)],verbose=False)
    predicted_values = model.predict(test_X)
    mae = mean_absolute_error(test_Y,predicted_values )
    return mae
```

```python
for estimators in [100,500,1000,1500]:
    mae = getMaeforXGBoost(estimators,train_X,test_X, train_y, test_y)
    print("Estimator : %d \t\t Mae : %d" %(estimators,mae))
```
